# data augmentation test

```python
import os
import joblib
import matplotlib.pyplot as plt
import numpy as np
data_dir = '/data2/obin/loc_data_12view_gibson/val'
data_list = [os.path.join(data_dir, x) for x in os.listdir(data_dir)]
```

```python
data = joblib.load(data_list[0])
```

```python
plt.imshow(data['rgb'])
plt.show()
```

```python
21 * 12
```

```python
data_patches = np.stack([data['rgb'][:,i*21:(i+1)*21] for i in range(12)])
index_list = np.arange(0,12).tolist()
random_cut = np.random.randint(12)
index_list = index_list[random_cut:] + index_list[:random_cut]
permuted_patches = data_patches[index_list]
augmented_img = np.concatenate(np.split(permuted_patches,12,axis=0),2)[0]
plt.imshow(augmented_img)
plt.show()
```
